```python
import pandas as pd
```

```python
test_1= pd.read_csv('Казань подвыборка_1 проверка.csv', sep=',', index_col=None)
test_2= pd.read_csv('Казань подвыборка_2 проверка.csv', sep=',', index_col=None)
test_3= pd.read_csv('Казань подвыборка_3 проверка.csv', sep=',', index_col=None)
test_4= pd.read_csv('Казань подвыборка_4 проверка.csv', sep=',', index_col=None)
test_5= pd.read_csv('Казань подвыборка_5 проверка.csv', sep=',', index_col=None)
test_1.columns=['id','подвыборка_1']
test_2.columns=['id','подвыборка_2']
test_3.columns=['id','подвыборка_3']
test_4.columns=['id','подвыборка_4']
test_5.columns=['id','подвыборка_5']
```

```python
test_all=test_1.merge(test_2,how='left',on='id')
test_all=test_all.merge(test_3,how='left',on='id')
test_all=test_all.merge(test_4,how='left',on='id')
test_all=test_all.merge(test_5,how='left',on='id')
test_all
```

```python
test_all.loc[(test_all['подвыборка_1']!=test_all['подвыборка_2'])|(test_all['подвыборка_2']!=test_all['подвыборка_3'])|(test_all['подвыборка_3']!=test_all['подвыборка_4'])|(test_all['подвыборка_4']!=test_all['подвыборка_5'])]
```

```python
test_all['crop']=test_all.mode(axis=1)[0].astype(int)
```

```python
test_all
```

```python
test_all['crop'].value_counts()
```

```python
test_all=test_all[['id','crop']]
```
